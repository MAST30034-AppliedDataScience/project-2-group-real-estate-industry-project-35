```python
import pandas as pd

# Load the data from both files using raw strings
property_data = pd.read_csv(r"../data/curated/property/property 3.csv")
merged_data = pd.read_csv(r"../data/curated/property/merged_property_sa2_data.csv")

# Merge the two dataframes on the 'address' column and keep the specified columns
merged_df = pd.merge(property_data, merged_data, on='address', how='inner')[
    ['price', 'address', 'beds', 'baths', 'parking', 'SA2_CODE_2021', 'SA2_NAME_2021', 'postcode']
]

# Save the merged dataframe to a new CSV file
merged_df.to_csv(r"../data/curated/external/unique_merged_data.csv", index=False)
```

```python
import pandas as pd

# Load the required data files
unique_merged_data = pd.read_csv(r"../data/curated/external/unique_merged_data.csv")
closest_station_data = pd.read_csv(r"../data/raw/external/API/3/closest_station_distance_final_3.csv")
closest_school_data = pd.read_csv(r"../data/raw/external/API/3/closest_school_distance_final_3.csv")


# Merge the station distance data into the unique_merged_data
unique_merged_data = pd.merge(
    unique_merged_data, 
    closest_station_data[['rent_address', 'min_driving_distance_km']], 
    left_on='address', 
    right_on='rent_address', 
    how='left'
)
unique_merged_data.rename(columns={'min_driving_distance_km': 'minimum_distance_station'}, inplace=True)

# Merge the school distance data into the updated unique_merged_data
unique_merged_data = pd.merge(
    unique_merged_data, 
    closest_school_data[['rent_address', 'min_driving_distance_km']], 
    left_on='address', 
    right_on='rent_address', 
    how='left'
)
unique_merged_data.rename(columns={'min_driving_distance_km': 'minimum_distance_school'}, inplace=True)

# Drop the extra 'rent_address' columns after merging
unique_merged_data.drop(columns=['rent_address_x', 'rent_address_y'], inplace=True)

# Save the changes by overwriting the original unique_merged_data.csv
unique_merged_data.to_csv(r"../data/curated/external/unique_merged_data.csv", index=False)

# Display the updated dataframe
print(unique_merged_data.head())
```

# crime

```python
import pandas as pd

# Load the required data files
unique_merged_data = pd.read_csv(r"../data/curated/external/unique_merged_data.csv")
crime_data = pd.read_csv(r"../data/curated/crime_by_year.csv/LGA_crime_postcode.csv")

# Filter crime data by year
crime_data_2021 = crime_data[crime_data['Year'] == 2021][['Postcode', 'Offence Count']]
crime_data_2022 = crime_data[crime_data['Year'] == 2022][['Postcode', 'Offence Count']]
crime_data_2023 = crime_data[crime_data['Year'] == 2023][['Postcode', 'Offence Count']]

# Merge the crime data for each year
unique_merged_data = pd.merge(
    unique_merged_data,
    crime_data_2021,
    left_on='postcode',
    right_on='Postcode',
    how='left'
)
unique_merged_data.rename(columns={'Offence Count': '2021crime'}, inplace=True)

unique_merged_data = pd.merge(
    unique_merged_data,
    crime_data_2022,
    left_on='postcode',
    right_on='Postcode',
    how='left'
)
unique_merged_data.rename(columns={'Offence Count': '2022crime'}, inplace=True)

unique_merged_data = pd.merge(
    unique_merged_data,
    crime_data_2023,
    left_on='postcode',
    right_on='Postcode',
    how='left'
)
unique_merged_data.rename(columns={'Offence Count': '2023crime'}, inplace=True)

# Drop the extra 'Postcode' columns after merging
unique_merged_data.drop(columns=['Postcode_x', 'Postcode_y', 'Postcode'], inplace=True)

# Save the changes by overwriting the original unique_merged_data.csv
unique_merged_data.to_csv(r"../data/curated/external/unique_merged_data.csv", index=False)

# Display the updated dataframe
print(unique_merged_data.head())
```
